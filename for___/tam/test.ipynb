```python
import hmac
import hashlib
import json

def generate_signature(payload: dict, secret_key: str="WEDDING_SECRET_SALT_2025") -> str:
    """
    Generates an HMAC-SHA256 signature for a JSON payload using a secret key.
    
    Args:
        payload: The data payload as a dictionary.
        secret_key: The secret key provided by the service.
    
    Returns:
        The generated signature as a hexadecimal string.
    """
    json_str = json.dumps(payload, separators=(',', ':'), ensure_ascii=False)
    
    msg = json_str
    
    # 文字列をバイト列にエンコード (UTF-8)
    msg_bytes = msg.encode('utf-8')
    
    # SHA-256ハッシュを計算
    hash_object = hashlib.sha256(msg_bytes)
    
    # 16進数文字列として返す
    return hash_object.hexdigest()

# --- Example Usage ---
# The payload to be signed
my_payload = {
    "action": "game_score",
    "timestamp": datetime(2025, 12, 12, 18, 0, 0).isoformat(),
    "name": "ashitayaru",
    "score": 65535,
    "clientId": 'unknown',
    "recaptchaToken": ''
}


# Generate the signature
api_signature = generate_signature(my_payload)
print(f"Generated Signature: {api_signature}")

# The generated signature can then be sent in an HTTP header, such as "X-Signature" 
# or "signature", as specified by the API documentation.
```

```python
import random 
today = datetime.now()
score = 76



def timedelta_to_iso(td):
    # timedeltaから合計秒数を取得
    total_seconds = int(td.total_seconds())
    
    # 日、時間、分、秒を計算
    days = total_seconds // 86400
    remaining_seconds = total_seconds % 86400
    hours = remaining_seconds // 3600
    remaining_seconds %= 3600
    minutes = remaining_seconds // 60
    seconds = remaining_seconds % 60

    # ISO 8601 期間形式 (例: P1DT2H30M5S) に整形
    return f"P{days}DT{hours}H{minutes}M{seconds}S"

params = {
    "action": "game_score",
    "timestamp": today.isoformat(),
    "name": "ashitayaru",
    "score": score,
    "clientId": 'unknown',
    "recaptchaToken": '',
    "gameToken": gameToken,
    "tapHistory": sorted([timedelta_to_iso(today-start+timedelta(0,0,int(random.random()*5),int(random.random()*1000))) for _ in range(score)])
}
# payload["signature"] = generate_signature(payload)

headers = { 'Content-Type':'application/x-www-form-urlencoded;charset=UTF-8' }

try:
    response = requests.post(url, data=urllib.parse.urlencode(params), headers=headers)
    # Check the response
    response.raise_for_status() # Raise an exception for bad status codes (4xx or 5xx)
    print("Request successful. Response content:")
    print(response.json()) # Access the response body as JSON
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
        
```
